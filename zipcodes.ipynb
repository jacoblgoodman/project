```python
import pandas as pd
import numpy as np
from uszipcode import SearchEngine, SimpleZipcode, Zipcode

```

```python
df = pd.read_feather('locations')
df['tuples'] = df.tuples.apply(lambda x: np.NaN if x == 'other' else eval(x))
df.head()
```

```python

lat = df.tuples[10][0]
long = df.tuples[10][1]
search = SearchEngine(simple_zipcode=False)
zip = search.by_coordinates(lat,long,radius = 25, returns =1)
zipcode = zip[0]
zipcode
```
