```python
def get_nulls_dtypes(df):
    """takes a data frame and returns by collumn the datat type,number of null values, 
    and the percentageof nulls in the original dataframe
    """
    # getting null values
    s_nullcount = df.isna().sum()
    #getting data types
    s_dtypes = df.dtypes
    #getting percents 
    s_percent = (s_nullcount/df.shape[0]).round(4)
    
    #merging
    df_nullDtype = pd.concat([s_dtypes,s_nullcount,s_percent], axis=1)
    #nameing collumns and index
    df_nullDtype.columns = ['datatypes', 'nulls', 'percent']
    df_nullDtype.index.name = 'columns'
    
    return df_nullDtype.sort_values(by = 'nulls',ascending =False)

df_summary = get_nulls_dtypes(df)
```
